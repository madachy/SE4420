```python
#Problem 1: The length of a queue during a 30 minute period is 3 for the first 5 minutes, 4 for the
# next 10 minutes, 2 for the next 6 minutes, and 3 for the last 9 minutes. What is the average
# queue length during this time?

# Queue length and time (minutes)
queue_length = [3, 4, 2, 3]
time_minutes = [5, 10, 6, 9]
#Matrix results in [15, 40, 12, 27]
# Compute weighted average
sum_of_time = sum(q * t for q, t in zip(queue_length, time_minutes))
total_time = sum(time_minutes)

average_queue_length = sum_of_time / total_time
average_queue_length
#equation equals 94/30
```
